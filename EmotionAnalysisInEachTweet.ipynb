```
# analyze each tweet
import datetime
import langdetect
from emotion_predictor import EmotionPredictor

tweet_fr = "Dans cette espace Bio, j'avoue être vraiment moi, authentique et pure."
tweet = "I believe what @GaryMarcusis highlighting is to consider #DL as one tool among others & suggests DL must be supplemented by other techniques..that I agree with somehow#AIDebate (1/2)"

# check whether tweet's language is English
lang = langdetect.detect(tweet)
if(lang != 'en'):
  print("I Cant Do That!")
else:
  print("Yeah, English")


t1 =  datetime.datetime.now()
# Emotion prediction
model = EmotionPredictor(classification='ekman', setting='mc')

t2 = datetime.datetime.now()
prediction = model.predict_classes([tweet])
t3 = datetime.datetime.now()
print(prediction)

print(t2 - t1)
print(t3 - t2)

# probability = model.predict_probabilities([tweet])
# print(probability)


```
