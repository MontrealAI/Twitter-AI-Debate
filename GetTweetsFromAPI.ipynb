```
# Get data from the Twitter using API
import tweepy
import json
import csv

# write your credential info of twitter's official API to a json file named 'cred_info.json'
with open('/content/gdrive/My Drive/Colab Notebooks/Twitter_Analysis/cred_info.json') as cred_json_file:
  credential_data = json.load(cred_json_file)

# credential info
consumer_key = credential_data['CONSUMER_KEY']
consumer_secret = credential_data['CONSUMER_SECRET']
access_key = credential_data['ACCESS_KEY']
access_secret = credential_data['ACCESS_SECRET']



# Create the api endpoint
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
api = tweepy.API(auth)

# maximum_number_of_tweets_to_be_extracted = 10 # how many tweets?
hashtag = 'AIDebate' # what hashtag?

# set the time range (twitter's API gives you data only since the recent week, therefore no matter how far you set the start time!)
since_date = "2019-12-16"
until_date = "2019-12-30"

cnt = 0
for tweet in tweepy.Cursor(api.search, q='#' + hashtag, until=until_date).items(): #search twitter for the hashtag with the desired criteria
  with open('/content/gdrive/My Drive/Colab Notebooks/Twitter_Analysis/tweets_with_hashtag_' + hashtag + '.txt', 'a') as the_file: #save as a file in google drive
    print(str(cnt) + '\n')
    the_file.write(str(tweet._json) + '\n') # json format
    cnt += 1

print ('Extracted ' + str(cnt) + ' tweets with hashtag #' + hashtag) # number of collected data
```
